```python
import pandas as pd
```

```python
data=pd.read_csv('data_m.csv')
```

```python
data
```

# Fill all NaN values with 0s

```python
data.fillna(111)
```

# Fill all NaN elements in columns 'Marks' and 'Grades' With 97 and 'AA' respectively.

```python
data
```

```python
data.fillna({'Marks':97,'Grades':'AA'})
```

```python
data
```

# Propagate non-null values forward

```python
data.fillna(method='ffill')
```

# Propagate non-null values backward.

```python
data.fillna(method='bfill')
```

# Axis

```python
data.fillna(method='bfill',axis='columns')
```

# Fill the first NaN element.

```python
data.fillna(method='ffill',limit=1)
```

```python
data
```

# Drop the rows where at least one value is not missing

```python
data.dropna(axis='index',how='any')
```

# Drop the columns where at least one value is missing

```python
data.dropna(axis=1,how='any')
```

# Drop the rows where all values are missing

```python
data.dropna(axis='index',how='all')
```

# Drop the columns where all values are missing

```python
data.dropna(axis='columns',how='all')
```

# Keep only the rows with at least 2 non-NA values

```python
data.dropna(axis=0,thresh=2)
```

# Define in which columns to look for missing/NA values

```python
data.dropna(subset=['Name','Grades'])
```

```python
data
```

```python
df1=data.dropna(inplace=False)
```
