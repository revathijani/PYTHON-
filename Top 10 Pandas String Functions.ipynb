```python
import pandas as pd
```

```python
dict1 = {'Name': ['Priyang Bhatt', 'Rashmin Panchal', 'Anil Rana', 'Rahul Patel'],
        'City': [' New York ', ' Los Angeles ', ' Chicago', ' Houston'],
        'State': ['NY', 'CA', 'IL', 'TX']}
data = pd.DataFrame(dict1)
```

```python
data
```

# 1. str.lower() and str.upper()

```python
data['name_lower']=data['Name'].str.lower()
```

```python
data
```

```python
data['name_upper']=data['Name'].str.upper()
```

```python
data
```

# 2.str.len()

```python
data['Name_len']=data['Name'].str.len()
```

```python
data
```

# 3. str.strip()

```python
data['City'][0]
```

```python
data['City_strip']=data['City'].str.strip()
```

```python
data['City_strip'][0]
```

```python
data['city_lstrip']=data['City'].str.lstrip()   #left strip
```

```python
data['city_lstrip'][0]
```

```python
data['city_rstrip']=data['City'].str.rstrip()    #Right Strip
```

```python
data['city_rstrip'][0]
```
